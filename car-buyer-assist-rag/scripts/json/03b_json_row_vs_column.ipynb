```python
import json
```

```python
# Row-oriented: Array of objects
row_format = [
    {"model": "Corolla", "type": "Sedan", "price": 22000},
    {"model": "RAV4", "type": "SUV", "price": 28000},
    {"model": "Camry", "type": "Sedan", "price": 26000}
]

print("Row-oriented format:")
print(json.dumps(row_format, indent=2))
```

```python
# Column-oriented: Object with arrays
column_format = {
    "model": ["Corolla", "RAV4", "Camry"],
    "type": ["Sedan", "SUV", "Sedan"],
    "price": [22000, 28000, 26000]
}

print("Column-oriented format:")
print(json.dumps(column_format, indent=2))
```

```python
# Convert row to column format
def row_to_column(row_data):
    if not row_data:
        return {}
    
    # Get all keys from first object
    keys = row_data[0].keys()
    
    # Build column format
    column_data = {key: [] for key in keys}
    
    for row in row_data:
        for key in keys:
            column_data[key].append(row[key])
    
    return column_data
```

```python
# Convert
column_result = row_to_column(row_format)
```

```python
print(json.dumps(column_result, indent=2))
```
