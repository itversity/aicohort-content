```python
import json
```

```python
# Array of car objects in single JSON file
cars_array = [
    {"model": "Toyota Corolla", "type": "Sedan", "price": 22000},
    {"model": "Toyota RAV4", "type": "SUV", "price": 30000},
    {"model": "Toyota Camry", "type": "Sedan", "price": 28000}
]
```

```python
cars_array[0]
```

```python

# Save as single JSON file
with open("cars_single.json", "w") as f:
    json.dump(cars_array, f, indent=2)

print("Created: cars_single.json")
```

```python
cars_list = [
    {"model": "Toyota Corolla", "type": "Sedan", "price": 22000},
    {"model": "Toyota RAV4", "type": "SUV", "price": 30000},
    {"model": "Toyota Camry", "type": "Sedan", "price": 28000}
]
```

```python
cars_list[0]
```

```python
# Save as JSONL - write each object on its own line
with open("cars_jsonl.jsonl", "w") as f:
    for car in cars_list:
        f.write(json.dumps(car) + "\n")

print("Created: cars_jsonl.jsonl")
```

```python
# Read JSONL file
cars_from_jsonl = []
```

```python

with open("cars_jsonl.jsonl", "r") as f:
    for line in f:
        car = json.loads(line.strip())
        cars_from_jsonl.append(car)

print(f"Loaded {len(cars_from_jsonl)} cars from JSONL\n")
```

```python
print(cars_from_jsonl[0])
```

```python
print(cars_from_jsonl)
```

```python

for car in cars_from_jsonl:
    print(f"{car['model']}: ${car['price']:,}")
```

```python
# Read single JSON file
with open("cars_single.json", "r") as f:
    cars = json.load(f)
```

```python
type(cars)
```

```python
cars[0]
```

```python
# Convert to JSONL
with open("converted_to_jsonl.jsonl", "w") as f:
    for car in cars:
        f.write(json.dumps(car) + "\n")

print("Converted JSON to JSONL")
```

```python
# Now convert JSONL to JSON
cars_from_jsonl = []
```

```python
with open("converted_to_jsonl.jsonl", "r") as f:
    for line in f:
        cars_from_jsonl.append(json.loads(line.strip()))
```

```python
cars_from_jsonl[0]
```

```python
# Save as single JSON
with open("converted_to_json.json", "w") as f:
    json.dump(cars_from_jsonl, f, indent=2)

print("Converted JSONL to JSON")
```

```python
# Save in both formats
# Single JSON
with open("toyota_specs.json", "w") as f:
    json.dump(toyota_models, f, indent=2)

print("  - toyota_specs.json (single file)")
```
