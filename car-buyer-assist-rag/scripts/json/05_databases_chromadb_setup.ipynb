```python
import chromadb
```

```python
print(f"ChromaDB version: {chromadb.__version__}")
print("Installation successful")
```

```python
client = chromadb.Client()
```

```python
collection = client.create_collection("test_cars")
```

```python
# Add test document
collection.add(
    documents=["Toyota Corolla: Affordable reliable sedan"],
    ids=["corolla_test"],
    metadatas=[{"model": "Corolla", "price": 22000}]
)
```

```python
print(f"Documents: {collection.count()}")
```

```python
# Query test
results = collection.query(
    query_texts=["economical car"],
    n_results=1
)
```

```python
print("Query: 'economical car'")
print(f"Found: {results['metadatas'][0][0]['model']}")
print("ChromaDB working!")
```
